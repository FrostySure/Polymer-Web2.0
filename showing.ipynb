```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 从文件加载数据
data = np.loadtxt('all-H2O-xyz.txt')

# 获取原子坐标数据
atom_x = data[:, 0]
atom_y = data[:, 1]
atom_z = data[:, 2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 画线
ax.plot(atom_x, atom_y, atom_z, c='black')

# 设置图形属性
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('Particle Trajectory')

plt.show()
```

```python
import numpy as np
import MDAnalysis as mda
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 读取DCD文件
u = mda.Universe("last.data", "msd.dcd")

# 获取特定类型原子的坐标数据
atom_x = []
atom_y = []
atom_z = []

target_type = 127  # 替换成你需要的目标类型
# target_id = 197936 #197924 197936 197939 197942 197915 197945 197921
# target_id = 197924 #197924 197936 197939 197942 197915 197945 197921
# target_id = 197939 #197924 197936 197939 197942 197915 197945 197921
# target_id = 197942 #197924 197936 197939 197942 197915 197945 197921
# target_id = 197915 #197924 197936 197939 197942 197915 197945 197921
target_id = 197945 #197924 197936 197939 197942 197915 197945 197921
# target_id = 197921 #197924 197936 197939 197942 197915 197945 197921

for ts in u.trajectory:
    coordinates = u.atoms.positions
#     selected_coordinates = coordinates[u.select_atoms(f"type {target_type}").indices]
    selected_coordinates = coordinates[u.select_atoms(f"id {target_id}").indices]
    
    atom_x.append(selected_coordinates[:, 0])
    atom_y.append(selected_coordinates[:, 1])
    atom_z.append(selected_coordinates[:, 2])

atom_x = np.array(atom_x)
atom_y = np.array(atom_y)
atom_z = np.array(atom_z)

# 画三维轨迹图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 画线
ax.plot(atom_x.flatten(), atom_y.flatten(), atom_z.flatten(), c='black')

# 设置图形属性
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title(f'Particle Trajectory (Type {target_type})')
ax.set_title(f'Particle Trajectory (ID {target_id})')

plt.show()
```
